# Graph of years

```python
import matplotlib.pyplot as plt
from collections import Counter

years = [
    2023, 2023, 2022, 2018, 2023, 2023, 2022, 2020, 2019, 2021, 2023, 2019, 2019, 2018, 2018, 2016, 2021, 2021, 2019, 2016,
    2019, 2017, 2023, 2020, 2021, 2023, 2023, 2018, 2017, 2023, 2021, 2023, 2022, 2020, 2020, 2017, 2022, 2017, 2022, 2019,
    2019, 2023, 2018, 2021, 2020, 2023, 2020, 2018, 2023, 2019, 2018, 2020, 2022, 2023, 2022, 2020, 2022, 2023, 2023, 2020,
    2020, 2023, 2022, 2023, 2022, 2022, 2019, 2018, 2021, 2022, 2022, 2023
]

# Count the occurrence of each year using Counter from the collections module
year_counts = Counter(years)

# Sorting the years for the graph to have a chronological order
sorted_years = sorted(year_counts.items())

# Splitting the sorted years and their counts into two lists for plotting
years, counts = zip(*sorted_years)

# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, counts, color='darkorange')  # You can change the color if you want
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Frequency', fontsize = 14)
plt.title('Frequency of Years', fontsize = 16)


plt.xticks(years, rotation=45)  # Rotate the x-axis labels for better readability

# Determine max count for setting y-ticks
max_count = max(counts)
# Set y-ticks to only whole numbers
plt.yticks(range(0, max_count + 2, max((max_count + 1) // 10, 1)))  # Adjust the step as needed

# Enable grid lines with customization for better readability
plt.grid(axis = 'y', color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
plt.gca().set_axisbelow(True)# puts the lines behind the bars--less tacky

plt.xticks(fontsize=14)                           # Makes the x-axis tick labels font size 14
plt.yticks(fontsize=14) 


plt.show()
```
